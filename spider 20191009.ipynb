```python
img_urls = []
for link in img:
    a = link.get('src')
    if a:
        if a.find('https://images') > -1:
            img_urls.append(a)
```

```python
img01 = requests.get(url=img_urls[0])
```

```python
with open('01.jpg','ab+') as f:
    for chunk in img01.iter_content(chunk_size = 1024):
        if chunk:
            f.write(chunk)
            f.flush()
```
