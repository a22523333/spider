```python
import matplotlib.pyplot as plt
import os
from time import time
import requests
from bs4 import BeautifulSoup
from queue import Queue
from threading import Thread
from io import BytesIO
from concurrent.futures import ThreadPoolExecutor
from multiprocessing import Process,Pool
```

```python
class DownloadBiaoQingBao(Thread):
    
    def __init__(self,queue,path):
        Thread.__init__(self)
        self.queue = queue
        self.path = path
        if not os.path.exists(path):
            os.makedirs(path)
            
    def run(self):
        while True:
            title,url = self.queue.get()
            try:
                download(title,url)
            finally:
                self.queue.task_done()
    
    def download(title,url):
        with open('./biaoqingbao/{title}.jpg'.format(title=title),'wb') as f:
            img = requests.get(url).content
            f.write(img)
```

```python
class DownloadBiaoqingbao(Thread):

    def __init__(self, queue, path):
        Thread.__init__(self)
        self.queue = queue
        self.path = './gongxiang/biaoqingbao/'
        if not os.path.exists(path):
            os.makedirs(path)

    def run(self):
        while True:
            url = self.queue.get()
            try:
                # print(url)
                download_biaoqingbaos(url, self.path)
            finally:
                self.queue.task_done()


def download_biaoqingbaos(url, path):

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    img_list = soup.find_all('img', class_='ui image lazy')

    for img in img_list:
        image = img.get('data-original')
        title = img.get('title')
        #print('下载图片： ', title)

        try:
            with open(path + title + os.path.splitext(image)[-1], 'wb') as f:
                img = requests.get(image).content
                f.write(img)
        except OSError:
            print('length  failed')
            break


if __name__ == '__main__':

    start = time()

    # 构建所有的链接
    _url = 'https://fabiaoqing.com/biaoqing/lists/page/{page}.html'
    urls = [_url.format(page=page) for page in range(1, 11)]

    queue = Queue()
    path = './gongxiang/biaoqingbao/'

    # 创建线程
    for x in range(10):
        worker = DownloadBiaoqingbao(queue, path)
        worker.daemon = True
        worker.start()

    # 加入队列
    for url in urls:
        queue.put(url)

    queue.join()

    print('下载完毕耗时：  ', time()-start)
```

```python
url = 'https://fabiaoqing.com/biaoqing/lists/page/{page}.html'
html = requests.get(url)
bf = BeautifulSoup(html.text)
```

```python
data = bf.find_all('img',class_='ui image lazy')
```

```python
dd = {}
for i in data:
    dd[i['title']] = i.get('data-original')
```

```python
def download(title,url):
    with open('./gongxiang/biaoqingbao/{title}.jpg'.format(title=title),'wb') as f:
        img = requests.get(url).content
        f.write(img)
```

```python
def download_biaoqingbaos(url, path):

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    img_list = soup.find_all('img', class_='ui image lazy')

    for img in img_list:
        image = img.get('data-original')
        title = img.get('title')
        #print('下载图片： ', title)

        try:
            with open(path + title + os.path.splitext(image)[-1], 'wb') as f:
                img = requests.get(image).content
                f.write(img)
        except OSError:
            print('length  failed')
            break
```

```python
url = 'https://fabiaoqing.com/biaoqing/lists/page/{page}.html'
path = './gongxiang/biaoqingbao/'
start = time()
with ThreadPoolExecutor(10) as e:
    for i in range(1,11):
        e.submit(download_biaoqingbaos,url.format(page=i),path)
print('下载完毕，耗时：',time()-start)
```
